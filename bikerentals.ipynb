# 11. Create a new categorical column called day_period, which can take four possible values: night, morning, afternoon and evening. These values correspond to the following binning of the hour column: 0-6: night, 6-12: morning, 12-6: afternoon, 6-24:evening.

```python
bin_edges = [0, 6, 12, 18, 24]  # Define the boundaries of each bin
bin_labels = ["night", "morning", "afternoon", "evening"]  # Labels for the bins

df['day_period'] = pd.cut(df['hour'], bins=bin_edges, labels=bin_labels, right=False)
```

```python
df.head(10)
```

# 12. Generate a pivot table for the mean of the hourly total rentals count, with the index set to the day period and the column set to the working day column. What can you observe from the table ?

```python

# Create the pivot table: index = day period, columns = workingday, values = mean count
pivot_table = df.pivot_table(values='count', index='day_period', columns='workingday', aggfunc='mean',)

# Print the pivot table
print(pivot_table)
```

```python
# Generate and print observations based on the pivot table
for day_period in pivot_table.index:
    workingday_value = pivot_table.loc[day_period, 1]
    non_workingday_value = pivot_table.loc[day_period, 0]
    
    if workingday_value > non_workingday_value:
        print(f"\nDuring {day_period}: Rentals are higher on working days ({workingday_value}) than on non-working days ({non_workingday_value}).")
    elif workingday_value < non_workingday_value:
        print(f"\nDuring {day_period}: Rentals are higher on non-working days ({non_workingday_value}) than on working days ({workingday_value}).")
    else:
        print(f"\nDuring {day_period}: Rentals are the same on working and non-working days ({workingday_value}).")
```

# 10. Plot the the mean  and the 95% confidence interval of the hourly total rentals count versus the period of the day column, which you created in the first part of the assignment. Which period of the day has the highest rentals count ? Does this peak period differ for working and non-working days ?

```python
# Grouping by 'day_period' and 'workingday' to calculate mean and 95% CI
grouped_workingday = df.groupby(['day_period', 'workingday'])['count']
mean_rentals_working = grouped_workingday.mean()
sem_rentals_working = grouped_workingday.sem()

# Calculate 95% confidence intervals (mean  1.96 * SEM)
ci_rentals_working = 1.96 * sem_rentals_working

# Plotting the results for working vs non-working days
fig, ax = plt.subplots(figsize=(10, 6))
for workingday in [0, 1]:
    period_rentals = mean_rentals_working.xs(workingday, level='workingday')
    ci = ci_rentals_working.xs(workingday, level='workingday')
    label = "Non-working days" if workingday == 0 else "Working days"
    ax.errorbar(period_rentals.index, period_rentals.values, yerr=ci.values, fmt='o', capsize=5, label=label)

ax.set_title('Mean Hourly Rentals Count with 95% Confidence Interval vs Period of the Day (Working vs Non-working Days)')
ax.set_xlabel('Period of the Day')
ax.set_ylabel('Mean Rentals Count')
ax.grid(True)
ax.legend()
plt.xticks(rotation=45)
plt.show()

# Identify the period with the highest rental count for working and non-working days
for workingday in [0, 1]:
    peak_period = mean_rentals_working.xs(workingday, level='workingday').idxmax()
    peak_value = mean_rentals_working.xs(workingday, level='workingday').max()
    print(f"For {'non-working' if workingday == 0 else 'working'} days, the peak rental period is '{peak_period}' with {peak_value} rentals.")
```
